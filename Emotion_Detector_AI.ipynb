```python
sentence = ["I got a new laptop today and I am excited"]

sentence_vector = vectorizer.transform(sentence)

prediction = model.predict(sentence_vector)

print("Emotion:", prediction[0])
```

```python
def detect_emotion(sentence):
    sentence_vector = vectorizer.transform([sentence])
    prediction = model.predict(sentence_vector)
    return prediction[0]
```

```python
detect_emotion("I passed my exam today")
```

```python
def detect_emotion_with_confidence(sentence):
    sentence_vector = vectorizer.transform([sentence])

    prediction = model.predict(sentence_vector)[0]

    probability = model.predict_proba(sentence_vector).max()

    confidence = round(probability * 100, 2)

    return prediction, confidence
```

```python
detect_emotion_with_confidence("I got a new laptop and I am excited")
```

```python
id2label = dataset["train"].features["label"].names

id2label
```

```python
train_data = dataset["train"]

texts = train_data["text"]
labels = train_data["label"]

print(len(texts))
print(len(labels))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)

X = vectorizer.fit_transform(texts)

y = labels

print(X.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X, y)

print("AI training complete ✅")
```

```python
def detect_emotion(sentence):
    sentence_vector = vectorizer.transform([sentence])

    prediction = model.predict(sentence_vector)[0]

    confidence = model.predict_proba(sentence_vector).max()

    emotion = id2label[prediction]

    return emotion, round(confidence * 100, 2)
```

```python
detect_emotion("I got a new laptop today and I am very excited")
```

```python
detect_emotion("I failed my exam and I feel terrible")
```

```python
detect_emotion("I am afraid of tomorrow")
```

```python
emotion_emoji = {
    "joy": "😊",
    "sadness": "😢",
    "anger": "😡",
    "fear": "😨",
    "love": "❤️",
    "surprise": "😲"
}


def emotion_app(sentence):
    emotion, confidence = detect_emotion(sentence)

    emoji = emotion_emoji[emotion]

    print(f"Emotion: {emoji} {emotion}")
    print(f"Confidence: {confidence}%")
```

```python
emotion_app("I got accepted into my dream university")
```
